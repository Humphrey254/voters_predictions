```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
%matplotlib inline

dataset1 = pd.read_csv("JoeBiden.csv")
dataset2 = pd.read_csv("JohnDelaney.csv")
dataset1
```

```python
# Combine the datasets
combined_dataset = pd.concat([dataset1, dataset2])

# Separate features (X) and target variable (y)
X = combined_dataset["Retweets"]
X = X.to_frame()  # Convert Series to DataFrame
X = combined_dataset["Retweets"].values.reshape(-1, 1)
y = combined_dataset["Mentions"]
```

```python
# Initialize and train the logistic regression model
model = LogisticRegression()
model.fit(X, y)
```

```python

# Make predictions for the combined dataset
predictions = model.predict(X)

```

```python
# Determine the winner based on predictions
winner = "Joe Biden" if (predictions == 1).sum() > (predictions == 0).sum() else "John Delaney"
print("Predicted Winner:", winner)
```

```python
import joblib

# save our logistic regression model
joblib.dump(model, 'logistic_voting_model.pkl')
```
